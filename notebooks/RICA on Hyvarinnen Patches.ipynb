Helper functions
================

```python
def scale_cols_to_unit_interval(arr):
    arr = arr.copy()
    arr -= arr.min(axis=0)
    arr /= arr.max(axis=0)
    arr *= 255
    return arr.astype('uint8')
```

Prepare Data
============

We load the Hyvarinnen patch data. It is available at Quoc Le's webpage with the original RICA code: http://cs.stanford.edu/~quocle/publications.html.

```python
datafile = 'hyv_patches_16.mat'
# Load data.                                                                                                   

f = h5py.File(datafile)
patches = np.array(f['patches'])

X = patches[:20000]
VX = patches[20000:30000]
TX = patches[30000:]

image_dims = 16, 16

X, VX, TX = [breze.learn.base.cast_array_to_local_type(i) for i in X, VX, TX]
```

Setup
=====

```python
n_features = 32
optimizer = 'rprop'
batch_size = X.shape[0]
m = rica.Rica(256, n_features, hidden_transfer='identity', c_ica=.05, fit_project_weights=False, batch_size=batch_size, out_transfer='identity', optimizer=optimizer, loss='squared')
losses = []

climin.initialize.randomize_normal(m.parameters.data, 0, .1)
```

Training
========

```python
max_passes = 50
batches_per_pass = X.shape[0] / batch_size
max_iter = batches_per_pass * max_passes

print "max_iter", max_iter

stop = climin.stops.after_n_iterations(max_iter)
pause = climin.stops.modulo_n_iterations(batches_per_pass)

# Set up a nice printout.
keys = '#', 'loss', 'val loss'#, 'step_length'
max_len = max(len(i) for i in keys)
header = '   '.join(i.ljust(max_len) for i in keys)
print header
print '-' * len(header)


for i, info in enumerate(m.powerfit((X,), (VX,), stop, pause)):
    losses.append((info['loss'], info['val_loss']))
    
    row = '%i' % i, '%.6f' % info['loss'], '%.6f' % info['val_loss'],# '%.6f' % info['step_length']
    print '   '.join(i.ljust(max_len) for i in row)
```
